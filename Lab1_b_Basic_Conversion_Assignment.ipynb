Q2. Read an color image, and display the image using plt.imshow after performing each of the following operations.

b. cv2.RGB2GRAY
c. cv2.BGR2GRAY

Q4. Read a color image as RGB. (which means read either convert cv2.BGR2RGB, since opencv reads image in BGR color scheme)

1. Resize it to 10x10, then in the RGB image; which has 3 channels, reduce the channel 1 and 2 to zero and then display the image.
2. Print the image before and after the transformation.




```python
if img is None:
    print("Error: Unable to read the image.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Resize the image to 10x10
    img_resized = cv2.resize(img_rgb, (10, 10))

    # Set channels 1 (Green) and 2 (Blue) to zero
    img_transformed = img_resized.copy()
    img_transformed[:, :, 1] = 0  # Green channel
    img_transformed[:, :, 2] = 0  # Blue channel

    plt.figure(figsize=(10, 4))

    plt.subplot(1, 3, 1)
    plt.title("Original RGB Image")
    plt.imshow(img_rgb)
    plt.axis('off')

    
    plt.subplot(1, 3, 2)
    plt.title("Resized to 10x10")
    plt.imshow(img_resized)
    plt.axis('off')

    
    plt.subplot(1, 3, 3)
    plt.title("Transformed (Green and Blue Channels to Zero)")
    plt.imshow(img_transformed)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

### Colorspaces and conversions
Ref: https://docs.opencv.org/3.4/de/d25/imgproc_color_conversions.html#color_convert_bayer


 Read a color image, resize to 64x64 then convert it to following formats and display those using plt.imshow

1. BGR to HSV
2. BGR to LAB
3. BGR to YUV
4. BGR to CIE



```python
def display_image(image, title):
    plt.figure(figsize=(4, 4))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()


if img is None:
    print("Error image.")
else:
    
    img_resized = cv2.resize(img, (64, 64))

    # BGR to HSV
    img_hsv = cv2.cvtColor(img_resized, cv2.COLOR_BGR2HSV)
    display_image(img_hsv, "BGR to HSV")

    # BGR to LAB
    img_lab = cv2.cvtColor(img_resized, cv2.COLOR_BGR2LAB)
    display_image(img_lab, "BGR to LAB")

    # BGR to YUV
    img_yuv = cv2.cvtColor(img_resized, cv2.COLOR_BGR2YUV)
    display_image(img_yuv, "BGR to YUV")

    # BGR to CIE
    img_cie = cv2.cvtColor(img_resized, cv2.COLOR_BGR2XYZ)
    display_image(img_cie, "BGR to CIE")
```

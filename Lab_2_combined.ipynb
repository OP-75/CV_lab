# Addition - Grey scale

```python
path = "images/sr2.jpg"
img = cv2.imread(path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print(img.shape)
# print(img) # elements

width = 10
height = 10
dim = (width, height)

# resize image
resized = cv2.resize(img, dim)


print(resized.shape)
print(resized)

plt.imshow(resized,cmap='gray')


```

```python
k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(resized.shape, dtype="uint8") * k

# print(resized)

print(M)


add1 = cv2.add(resized, M)

add2 = resized + M

print("The addition with CV2", add1)

# overflow happens in this so 216+50 = 10 (caz integer overflow after 255)
print("The addition with simple operator", add2)


print("Shape of input image", resized.shape)


#plt.imshow(resized,cmap='gray')
#plt.show()

plt.imshow(resized, cmap='gray')
plt.show()


plt.imshow(add1, cmap='gray')
plt.show()

plt.imshow(add2,cmap='gray')
plt.show()

```

# Subtraction

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(img.shape, dtype="uint8") * k

sub = cv2.subtract(img, M)

plt.imshow(img)
plt.show()
plt.imshow(sub)
plt.show()
```

# Multiplication

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(img.shape, dtype="uint8") * k

mul = cv2.multiply(img, M)

plt.imshow(img)
plt.show()
plt.imshow(mul)
plt.show()
```

# Division

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(img.shape, dtype="uint8") * k

div = cv2.divide(img, M)

plt.imshow(img)
plt.show()
plt.imshow(div)
plt.show()
```

# AND

```python
blank = np.ones((400,400), dtype='uint8')

img1 = cv2.rectangle(blank.copy(),(30,30),(370,370), 0, -1)
img2 = cv2.circle(blank.copy(),(200,200), 200,0, -1)

print(img1.shape)
print(img2.shape)

# resixe 10x10
width = 400
height = 400
dim = (width, height)

# resize image
imgR1 = cv2.resize(img1, dim)
imgR2 = cv2.resize(img2, dim)

AND = cv2.bitwise_and(imgR1, imgR2)

plt.imshow(imgR1, cmap='gray')
plt.show()
print(imgR1)
plt.imshow(imgR2, cmap='gray')
plt.show()
print(imgR2)
plt.imshow(AND, cmap='gray')
plt.show()
print(AND)
```

# OR

```python
blank = np.ones((400,400), dtype='uint8')
img1 = cv2.rectangle(blank.copy(),(30,30),(370,370), 0, -1)
img2 = cv2.circle(blank.copy(),(200,200), 200,0, -1)

print(img1.shape)
print(img2.shape)

# resixe 10 x10
width = 400
height =400
dim = (width, height)

# resize image
imgR1 = cv2.resize(img1, dim)
imgR2 = cv2.resize(img2, dim)

OR = cv2.bitwise_or(imgR1, imgR2)

plt.imshow(imgR1, cmap='gray')
plt.show()
print(imgR1)
plt.imshow(imgR2, cmap='gray')
plt.show()
print(imgR2)
plt.imshow(OR, cmap='gray')
plt.show()
print(OR)
```

## Intensity Transformations

Perform the following operations
1. Image negative
Slicing without background
2. slicing with background
3. Log transform
4. Power Law

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imageio
import matplotlib.pyplot as plt
```

### Image negative

```python


path = r"C:\Users\hpary\OneDrive\Documents\CV\images\sr2.jpg"

image = cv2.imread(path)

plt.figure(figsize = (6,6))
plt.imshow(image, 'gray')
plt.figure(figsize = (6,6))
plt.imshow(255 - image);

```



# Intensity level slicing without background

```python
img = cv2.imread(path)

image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()

z = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        if(gray[i][j]>50 and gray[i][j]<150):
            z[i][j]=255
        else:
            z[i][j]=0



plt.imshow(z, 'gray')
plt.title('Sliced Original Image')
plt.show()

```

# Intensity level slicing with background

```python
image = cv2.imread(path)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()

z = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        if(gray[i][j]>50 and gray[i][j]<150):
            z[i][j]=255
        else:
            z[i][j]=gray[i][j]

plt.imshow(z, 'gray')
plt.title('Sliced with background')
plt.show()
```

# Log transformation

```python
image = cv2.imread(path)

image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()

c = 255/(np.log(1+np.max(gray)))

log = c*np.log(1+gray)
log = np.array(log,dtype=np.uint8)

plt.imshow(log, 'gray')
plt.show()
```

# Power Law

```python
image = cv2.imread(path)

image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()

z = np.zeros((x,y))

gamma = 0.2
c = 255/(np.log(1+np.max(gray)))
z = np.array(c*(gray**gamma), dtype = 'uint8')

plt.imshow(z, 'gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# AND

```python
blank = np.ones((400,400), dtype='uint8')

img1 = cv2.rectangle(blank.copy(),(30,30),(370,370), 0, -1)
img2 = cv2.circle(blank.copy(),(200,200), 200,0, -1)

print(img1.shape)
print(img2.shape)

# resixe 10x10
width = 400
height = 400
dim = (width, height)

# resize image
imgR1 = cv2.resize(img1, dim)
imgR2 = cv2.resize(img2, dim)

AND = cv2.bitwise_and(imgR1, imgR2)

plt.imshow(imgR1, cmap='gray')
plt.show()
print(imgR1)
plt.imshow(imgR2, cmap='gray')
plt.show()
print(imgR2)
plt.imshow(AND, cmap='gray')
plt.show()
print(AND)
```

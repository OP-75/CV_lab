```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("1. Addition\n2. Subtraction\n3. Multiplication\n4. Division\n")
```

# Addition - Grey scale

```python
path = "images/sr2.jpg"
img = cv2.imread(path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print(img.shape)
# print(img) # elements

width = 10
height = 10
dim = (width, height)

# resize image
resized = cv2.resize(img, dim)


print(resized.shape)
print(resized)

plt.imshow(resized,cmap='gray')


```

```python
k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(resized.shape, dtype="uint8") * k

# print(resized)

print(M)


add1 = cv2.add(resized, M)

add2 = resized + M

print("The addition with CV2", add1)

# overflow happens in this so 216+50 = 10 (caz integer overflow after 255)
print("The addition with simple operator", add2)


print("Shape of input image", resized.shape)


#plt.imshow(resized,cmap='gray')
#plt.show()

plt.imshow(resized, cmap='gray')
plt.show()


plt.imshow(add1, cmap='gray')
plt.show()

plt.imshow(add2,cmap='gray')
plt.show()

```

# Addition Color image

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


print(img.shape)
# print(img)


width = 10
height = 10
dim = (width, height)

# resize image
resized = cv2.resize(img, dim)

print(resized.shape)
print(resized)

plt.imshow(resized)


```

```python
k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(resized.shape, dtype="uint8") * k

print("OG img",resized)

print("M to add",M)


add1 = cv2.add(resized, M)

add2 = resized + M

print("The addition with CV2", add1)


print("The addition with simple operator", add2)

print("Shape of input image", resized.shape)

plt.imshow(resized)
plt.show()


plt.imshow(add1)
plt.show()

plt.imshow(add2)
plt.show()

```

# Subtraction

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(img.shape, dtype="uint8") * k

sub = cv2.subtract(img, M)

plt.imshow(img)
plt.show()
plt.imshow(sub)
plt.show()
```

# Multiplication

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(img.shape, dtype="uint8") * k

mul = cv2.multiply(img, M)

plt.imshow(img)
plt.show()
plt.imshow(mul)
plt.show()
```

# Division

```python
path = "images/sr2.jpg"
img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = int(input("Enter intensity of 2nd img pixels: "))
M = np.ones(img.shape, dtype="uint8") * k

div = cv2.divide(img, M)

plt.imshow(img)
plt.show()
plt.imshow(div)
plt.show()
```
